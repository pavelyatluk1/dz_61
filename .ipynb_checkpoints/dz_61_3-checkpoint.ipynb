#### 2. Використовуючи бібліотеку zip, напишіть код для стиснення текстового файлу.

```python
import gzip

with open('text_file.txt', 'rt') as file_in:
    with gzip.open('compressed_file.gz', 'wt') as file_out:
        file_out.writelines(file_in)
print('File compression complete.')
```

#### 3. Реалізуйте алгоритм Хаффмана для стиснення текстових даних. Напишіть код для стиснення та розпакування текстового файлу за допомогою цього алгоритму.

```python
import heapq
from heapq import heappop, heappush 
 
def isLeaf(root):
    return root.left is None and root.right is None 
 
 # вузол дерева
class Node:
    def __init__(self, ch, freq, left=None, right=None):
        self.ch = ch
        self.freq = freq
        self.left = left
        self.right = right
 
    # Заміна функції `__lt__()`, щоб клас `Node` працював із пріоритетною чергою
    # таким чином, що елемент з найвищим пріоритетом має найнижчу частоту
    def __lt__(self, other):
        return self.freq < other.freq 
 
 # Перейдіть по дереву Хаффмана та зберігайте коди Хаффмана в словнику
def encode(root, s, huffman_code):
 
    if root is None:
        return
 
    # знаходження листового вузла
    if isLeaf(root):
        huffman_code[root.ch] = s if len(s) > 0 else '1'
 
    encode(root.left, s + '0', huffman_code)
    encode(root.right, s + '1', huffman_code) 
 
 # Перейдіть по дереву Хаффмана та декодуйте закодований рядок
def decode(root, index, s):
 
    if root is None:
        return index
 
    # знаходження листового вузла
    if isLeaf(root):
        print(root.ch, end='')
        return index
 
    index = index + 1
    root = root.left if s[index] == '0' else root.right
    return decode(root, index, s) 
 
 # Будує дерево Хаффмана та декодує поданий вхідний текст
def buildHuffmanTree(text):
 
    # базовий регістр: порожній рядок
    if len(text) == 0:
        return
 
    # підрахувати частоту появи кожного персонажа і збережіть його в словнику
    freq = {i: text.count(i) for i in set(text)}
 
    # Створіть пріоритетну чергу для зберігання живих вузлів дерева Хаффмана
    pq = [Node(k, v) for k, v in freq.items()]
    heapq.heapify(pq)
 
    # робити, доки в черзі не залишиться більше одного вузла
    while len(pq) != 1:
 
        # Видаліть два вузли з найвищим пріоритетом (найменша частота) з черги 
        left = heappop(pq)
        right = heappop(pq) 

        # створити новий внутрішній вузол із цими двома вузлами як дочірніми та з частотою, що дорівнює сумі частот двох вузлів.
        # Додайте новий вузол до черги пріоритетів. 
        total = left.freq + right.freq
        heappush(pq, Node(None, total, left, right))
 
    root = pq[0]
 
    # пройти по дереву Хаффмана і зберегти коди Хаффмана в словнику
    huffmanCode = {}
    encode(root, '', huffmanCode)
 
    # надрукуйте коди Хаффмана
    print('Huffman Codes are:', huffmanCode)
    print('The original string is:', text)
 
    # надрукувати закодований рядок
    s = ''
    for c in text:
        s += huffmanCode.get(c)
 
    print('The encoded string is:', s)
    print('The decoded string is:', end=' ')
 
    if isLeaf(root):
        while root.freq > 0:
            print(root.ch, end='')
            root.freq = root.freq - 1
    else:
        index = -1
        while index < len(s) - 1:
            index = decode(root, index, s) 
 
 # Реалізація алгоритму кодування Хаффмана в Python
if __name__ == '__main__':
 
    text = 'Дуже цікавий алгоритм та його реалізація на Пайтон'
    buildHuffmanTree(text)
```
